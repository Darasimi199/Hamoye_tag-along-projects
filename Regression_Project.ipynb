```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
```

```python
df_2 = df.drop(columns=['date', 'lights'])

normalised_df = pd.DataFrame(scaler.fit_transform(df_2), columns = df_2.columns)
```

```python
normalised_df.head()
```

```python
features_df = normalised_df.drop(columns=['Appliances'])

App_target = normalised_df['Appliances']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(features_df,
                                                    App_target, test_size=0.3,
                                                    random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression()

linear_model.fit(x_train, y_train)
```

```python
predicted_values = linear_model.predict(x_test)
```

```python
predicted_values[:10]
```

```python
from sklearn.metrics import mean_absolute_error

mae =  mean_absolute_error(y_test, predicted_values)

round(mae,2)
```

```python
from sklearn.metrics import mean_squared_error

rmse = np.sqrt(mean_squared_error(y_test, predicted_values))

round(rmse, 3)
```

```python
from sklearn.metrics import r2_score

r_squared = r2_score(y_test, predicted_values)

round(r_squared, 2)
```

```python
from sklearn.linear_model import Ridge

ridge_reg = Ridge(alpha = 0.4)

ridge_reg.fit(x_train, y_train)
```

```python
pred_2 = ridge_reg.predict(x_test)

pred_2[:10]
```

```python
from sklearn.metrics import mean_squared_error

rmse = np.sqrt(mean_squared_error(y_test, pred_2))

round(rmse, 3)
```

```python
from sklearn.linear_model import Lasso

lasso_reg = Lasso(alpha = 0.001)

lasso_reg.fit(x_train, y_train)
```

```python
pred_3 = lasso_reg.predict(x_test)

pred_3[:10]
```

```python
from sklearn.metrics import mean_squared_error

rmse = np.sqrt(mean_squared_error(y_test, pred_3))

round(rmse, 3)
```

```python
def get_weight_df(model, feat, col_name):
  # this function returns the weight of every feature
  weights = pd.Series(model.coef_, feat.columns).sort_values()
  weights_df = pd.DataFrame(weights).reset_index()
  weights_df.columns = ['Features', col_name]
  weights_df[col_name].round(3)
  return weights_df

lasso_weights_df = get_weight_df(lasso_reg, x_train, 'Lasso_Weight')
```
