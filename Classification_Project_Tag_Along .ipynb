## Exploratory Data Analysis

```python
# How many target observations were stable and unstable

splot = sns.barplot(x=df['stabf'].value_counts().index, y=df['stabf'].value_counts())
for p in splot.patches:
    splot.annotate(format(p.get_height(), '.1f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 9), 
                   textcoords = 'offset points')
```

From that, unstable counts are almost twice the stable counts

```python
# Dropping the 'stab' column

df = df.drop('stab', axis=1)

df.head()
```

```python
# Splitting Dataset into target and feature
y = df['stabf']

X = df.drop('stabf', axis=1)
```

```python
X.head()
```

```python
# Split into 80-20 train and test data with random_state=1

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2,
                                                   random_state=1)
```

```python
# Normalize the x_train and x_test

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
norm_train_df = scaler.fit_transform(x_train)
norm_test_df = scaler.fit_transform(x_test)
```

## Decision Trees

```python
from sklearn.tree import DecisionTreeClassifier

dec_tree = DecisionTreeClassifier(random_state=1)
dec_tree.fit(norm_train_df, y_train)
```

```python
# Predictions for DT

dt_pred = dec_tree.predict(norm_test_df)

dt_pred[:15]
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(random_state=1)
rfc.fit(norm_train_df, y_train)
```

```python
rfc_pred = rfc.predict(norm_test_df)

rfc_pred[:15]
```

## ExtraTreesClassifier

```python
from sklearn.ensemble import ExtraTreesClassifier

etc = ExtraTreesClassifier(random_state=1)
etc.fit(norm_train_df, y_train)
```

```python
etc.pred = etc.predict(norm_test_df)

etc.pred[:15]
```

## XGBoost

```python
import xgboost as xgb
```

```python
xgbc = xgb.XGBClassifier(random_state=1)
xgbc.fit(norm_train_df, y_train)
```

```python
xgbc_pred = xgbc.predict(norm_test_df)

xgbc_pred[:15]
```

## LightGBM

```python
import lightgbm as ltb
```

```python
ltb_class = ltb.LGBMClassifier(random_state=1)
```

```python
ltb_class.fit(norm_train_df, y_train)
```

```python
ltb_class_pred = ltb_class.predict(norm_test_df)

ltb_class_pred[:15]
```

```python
# Accuracy Score for Random Forest

from sklearn.metrics import (recall_score, accuracy_score,
                             precision_score, f1_score, confusion_matrix)

accuracy = accuracy_score(y_true = y_test, y_pred = rfc_pred)
accuracy
#print('Accuracy: {}'.format(round(accuracy*100), 4))
```

```python
# Accuracy score for XGBoost

acc = accuracy_score(y_true = y_test, y_pred = xgbc_pred)
acc
```

```python
#Accuracy score for LightGBM

acc_1 = accuracy_score(y_true = y_test, y_pred = ltb_class_pred)
acc_1
```
